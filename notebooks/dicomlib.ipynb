```python
from dicomlib import association, server

logger = logging.getLogger(__name__)

with server(('', 0)) as scp:
    scp_address = scp.server_address
    with association(scp_address) as scu:  # pyright: ignore[reportArgumentType]
        status = scu.send_c_echo()
        logger.info('Status of C-ECHO: %s', status)
```

```python
from pynetdicom.status import Status

assert status.Status == Status.SUCCESS
```

## `C-STORE`

Spawn a `storescp` server and send a DICOM image to it

```python
from pydicom.data import get_testdata_file

import dicomlib

dataset = get_testdata_file('CT_small.dcm', read=True)

with dicomlib.server(('', 0), evt_handlers=[(dicomlib.EVT_C_STORE, dicomlib.handlers.c_store_tmp_dir)]) as scp:
    with dicomlib.association(scp.server_address) as scu:
        scu.send_c_store(dataset)
```
